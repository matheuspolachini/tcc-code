```python
import os
import math
import string
import random
import shutil
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from tecnicas.pvd import embed_message_pvd
```

```python
base_dir = Path('/media/external/tcc_work')
base_dir.cwd()
```

# Cria conjuntos de treinamento/teste

```python
dataset_folder = base_dir / 'original_png'
images = [file for file in dataset_folder.iterdir()]
print(f'Dataset has {len(images)} images')
random.shuffle(images)

# 5% da capacidade do LSB
target_byte_count = math.floor((255 * 255 * 3) * 0.05)

target_path = base_dir / 'pvd'
if target_path.exists():
    shutil.rmtree(target_path)
target_path.mkdir()

images_len = len(images)
images_len_half = images_len // 2
for i, image_path in enumerate(images[:images_len_half]):
    print(f'Encoding image {i}')
    embed_str = ''.join(random.choice(string.ascii_letters) for x in range(target_byte_count))
    filename = image_path.stem
    output_path = target_path / f"encoded_{filename}.png"
    print(f"Calling with {image_path.as_posix()} and {output_path.as_posix()}")
    embed_message_pvd(image_path.as_posix(), embed_str, output_path.as_posix())

print(f"Copying remaining images")
for image_path in images[images_len_half:]:
    filename = image_path.stem
    output_path = target_path / f"{filename}.png"
    shutil.copy(image_path, output_path)
```
